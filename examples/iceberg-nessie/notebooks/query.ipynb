# Iceberg + Nessie (host)

Stack exposes **19120** (Nessie) and **9000** (MinIO). After the app commits data, list tables via the Nessie-backed Iceberg catalog.

If `api/v1` fails with your Iceberg/Nessie versions, try `http://127.0.0.1:19120/api/v2` in the cell below.

```python
from pyiceberg.catalog import load_catalog

catalog = load_catalog(
    "examples",
    type="nessie",
    uri="http://127.0.0.1:19120/api/v1",
    ref="main",
    **{
        "warehouse": "s3://warehouse/iceberg-nessie",
        "io-impl": "org.apache.iceberg.aws.s3.S3FileIO",
        "s3.endpoint": "http://127.0.0.1:9000",
        "s3.access-key-id": "admin",
        "s3.secret-access-key": "password",
        "s3.path-style-access": "true",
    },
)

list(catalog.list_namespaces())
```

```python
tid = ("postgres_public", "demo_orders")
if catalog.table_exists(tid):
    print(catalog.load_table(tid).scan().to_arrow())
else:
    print("Table not present yet — run scripts/load.sh while the app is up.")
```
