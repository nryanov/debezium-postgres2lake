# Paimon JDBC — warehouse objects (MinIO)

There is no PyIceberg-equivalent notebook for Paimon yet. This lists **S3** keys under the `warehouse` bucket to confirm data files landed under the `paimon-jdbc/` prefix.

Requires: `pip install boto3` (included in `../../common/requirements-python.txt`).

```python
import boto3

s3 = boto3.client(
    "s3",
    endpoint_url="http://127.0.0.1:9000",
    aws_access_key_id="admin",
    aws_secret_access_key="password",
    region_name="us-east-1",
)

prefix = "paimon-jdbc/"
resp = s3.list_objects_v2(Bucket="warehouse", Prefix=prefix, MaxKeys=50)
for obj in resp.get("Contents", []):
    print(obj["Key"])
```
