# Iceberg + Hive Metastore (host)

Thrift HMS is on **9083**. Warehouse in this example is `s3a://warehouse/hive-warehouse` (same as the Hive Metastore container).

First startup of `hive-metastore` can take several minutes while schema is initialized.

```python
from pyiceberg.catalog import load_catalog

catalog = load_catalog(
    "examples",
    type="hive",
    uri="thrift://127.0.0.1:9083",
    warehouse="s3://warehouse/hive-warehouse",
)

list(catalog.list_namespaces())
```

```python
tid = ("postgres_public", "demo_orders")
if catalog.table_exists(tid):
    print(catalog.load_table(tid).scan().to_arrow())
else:
    print("Table not present yet — run scripts/load.sh while the app is up.")
```
