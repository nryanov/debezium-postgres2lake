# Paimon Hive catalog — warehouse layout

Paimon files share the Hive warehouse prefix `hive-warehouse/` in this example. After loading events, you should see keys under `hive-warehouse/postgres_public.db/` (layout may vary by Paimon version).

```python
import boto3

s3 = boto3.client(
    "s3",
    endpoint_url="http://127.0.0.1:9000",
    aws_access_key_id="admin",
    aws_secret_access_key="password",
    region_name="us-east-1",
)

prefix = "hive-warehouse/"
resp = s3.list_objects_v2(Bucket="warehouse", Prefix=prefix, MaxKeys=50)
for obj in resp.get("Contents", []):
    print(obj["Key"])
```
