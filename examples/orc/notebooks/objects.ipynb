# ORC output under MinIO

List `.orc` objects under the unpartitioned layout for `demo_orders`.

```python
import boto3

s3 = boto3.client(
    "s3",
    endpoint_url="http://127.0.0.1:9000",
    aws_access_key_id="admin",
    aws_secret_access_key="password",
    region_name="us-east-1",
)

prefix = "postgres/public/demo_orders/"
resp = s3.list_objects_v2(Bucket="warehouse", Prefix=prefix, MaxKeys=30)
for obj in resp.get("Contents", []):
    if obj["Key"].endswith(".orc"):
        print(obj["Key"], obj["Size"])
```
